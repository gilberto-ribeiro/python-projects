```python
import pandas as pd
import numpy as np
```

```python
vendas_clientes = pd.read_json('https://caelum-online-public.s3.amazonaws.com/2928-transformacao-manipulacao-dados/dados_vendas_clientes.json')
vendas_clientes = pd.json_normalize(vendas_clientes.dados_vendas)
vendas_clientes.head()
```

```python
vendas_clientes = vendas_clientes.explode(column=['Cliente', 'Valor da compra'])
vendas_clientes['Valor da compra'] = vendas_clientes['Valor da compra'].apply(lambda x: x.replace('R$', '').replace(',', '.').strip())
vendas_clientes['Valor da compra'] = vendas_clientes['Valor da compra'].astype(np.float64)
vendas_clientes
```

```python
vendas_clientes['Cliente'] = vendas_clientes['Cliente'].str.replace('[^A-Za-záàâãéèêíïóôõöúçñÁÀÂÃÉÈÍÏÓÔÕÖÚÇÑ ]', '', regex=True)
vendas_clientes['Cliente'] = vendas_clientes['Cliente'].str.title()
vendas_clientes['Cliente'] = vendas_clientes['Cliente'].str.strip()
vendas_clientes
```
