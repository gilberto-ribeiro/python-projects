```python
import pandas as pd
import numpy as np
```

```python
locacao_imoveis = pd.read_json('https://caelum-online-public.s3.amazonaws.com/2928-transformacao-manipulacao-dados/dados_locacao_imoveis.json')
locacao_imoveis = pd.json_normalize(locacao_imoveis.dados_locacao)
locacao_imoveis.head()
```

```python
colunas = list(locacao_imoveis.columns)
locacao_imoveis = locacao_imoveis.explode(column=colunas[1:])
locacao_imoveis.valor_aluguel = locacao_imoveis.valor_aluguel.apply(lambda x: x.replace('$', '').replace('reais', '').replace(',', '.').strip()).astype(np.float64)
locacao_imoveis.head()
```

```python
locacao_imoveis['apartamento'] = locacao_imoveis['apartamento'].str.replace(' (blocoAP)', '')
locacao_imoveis.head()
```

```python
colunas_data = ['datas_combinadas_pagamento', 'datas_de_pagamento']
for coluna in colunas_data:
    locacao_imoveis[coluna] = pd.to_datetime(locacao_imoveis[coluna], format='%d/%m/%Y')
```

```python
locacao_imoveis = locacao_imoveis.assign(diferenca_de_tempo = lambda x: x.datas_de_pagamento - x.datas_combinadas_pagamento)
locacao_imoveis['diferenca_de_tempo'] = locacao_imoveis['diferenca_de_tempo'].dt.days
locacao_imoveis
```
