## Desafio: habitantes por país

```python
url_pais = 'https://pt.wikipedia.org/wiki/Lista_de_pa%C3%ADses_por_popula%C3%A7%C3%A3o'
dados_pais = pd.read_html(url_pais)[0]
```

```python
dados_pais.columns
```

```python
dados_pais.drop(columns=['Unnamed: 0', 'Estimativa Oficial'], inplace=True)
```

```python
dados_pais['Estimativa da ONU'] = dados_pais['Estimativa da ONU'].str.replace(' ', '')
```

```python
dados_pais['Estimativa da ONU'] = dados_pais['Estimativa da ONU'].astype(int)
```

```python
dados_pais[0:10].sort_index(ascending=True).plot(x='País (ou território dependente)', y='Estimativa da ONU', kind='bar')
```
