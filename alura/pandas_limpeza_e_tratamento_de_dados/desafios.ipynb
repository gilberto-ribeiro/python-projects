# Desafio 2

```python
dados_pessoas = pd.read_json('dados.json')
dados_pessoas = pd.json_normalize(dados_pessoas['pessoas'], sep=' - ')
dados_pessoas = dados_pessoas.explode('telefones')
dados_pessoas.reset_index(drop=True, inplace=True)
dados_pessoas.head()
```

```python
dados_pessoas['idade'] = dados_pessoas['idade'].astype(int)
indices_rua_vazia = dados_pessoas[dados_pessoas['endereco - rua'] == ''].index
dados_pessoas.loc[indices_rua_vazia, 'endereco - rua'] = 'Rua B'
dados_pessoas
```

# Desafio 3

```python
with open('cursos_cadastrados.json', 'r', encoding='utf-8') as arquivo_cursos:
    dados_cursos = json.load(arquivo_cursos)
```

```python
dados_cursos = pd.json_normalize(dados_cursos)
dados_cursos
```

```python
dados_cursos.dropna(inplace=True)
dados_cursos
```

```python
dados_cursos.drop_duplicates(inplace=True)
dados_cursos
```

```python
dados_cursos.replace('', pd.NA, inplace=True)
dados_cursos
```

```python
dados_cursos.dropna(inplace=True)
dados_cursos
```

```python
colunas_datas = ['data_inicio', 'data_conclusao']
colunas_numericas = ['concluintes', 'preco']
```

```python
for coluna in colunas_datas:
    dados_cursos[coluna] = pd.to_datetime(dados_cursos[coluna])
```

```python
for coluna in colunas_numericas:
    dados_cursos[coluna] = pd.to_numeric(dados_cursos[coluna])
```
